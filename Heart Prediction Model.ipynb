```python
clinical_records = pd.read_csv('../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
```

# Correlations

```python
corr_matrix = clinical_records.corr()
corr_matrix["DEATH_EVENT"].sort_values(ascending=False)
```

```python
y = clinical_records["DEATH_EVENT"]
```

```python
new_x =clinical_records[['age', 'ejection_fraction', 'serum_sodium', 'time']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(new_x, y, test_size = 0.2, random_state = 2698)
```

```python
X_train
```

# Models

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(max_iter = 1000)
log_reg.fit(X_train, y_train)
log_pred = log_reg.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, log_pred))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier(random_state=5)

forest_clf.fit(X_train, y_train)

forest_pred = forest_clf.predict(X_test)

accuracy_score(y_test, forest_pred)
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gradientboost_clf = GradientBoostingClassifier(max_depth = 1)
gradientboost_clf.fit(X_train,y_train)
gradientboost_pred = gradientboost_clf.predict(X_test)
accuracy_score(y_test,gradientboost_pred)
```

# SVM

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train,y_train)
svc_pred = svm.predict(X_test)
accuracy_score(y_test,svc_pred)
```

# AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
ada_clf = AdaBoostClassifier(
 DecisionTreeClassifier(max_depth=1), n_estimators=200,
 algorithm="SAMME.R", learning_rate=0.3)
ada_clf.fit(X_train, y_train)

ada_pred = ada_clf.predict(X_test)

accuracy_score(y_test, ada_pred)
```

# Bonus Tip:

If you change the random_state paramater when performing train_test_split, you actually get very different results. For example, when you change the random_state to random_state=1, the results are as follows:

Logistic Regression: 86.6%

Random Forest: 86.6%

Gradient Boost: 83.3%

SVM: 90%

AdaBoost: 85%

As you can see, when we change the random_state parameter, we get very different results which lead to worse performing models.

Another example, that really takes a toll on the accuracies is when I change the random_state paramater to random_state=42, which is what many practioners use as the random_state. The results are as follows:

Logistic Regression: 81.6%

Random Forest: 76.6%

Gradient Boost: 78.3%

SVM: 71.6%

AdaBoost: 71.6%

These results are not very appealing and this was only due to changing the random_state paramter. We don't get anywhere near the results that we did when random_state was equal to 1 or 2 and this was something I wanted to share and a cool discovery I had. I didn't realize that changing such a simple paramter could really be the difference between the highest accuracy of 93.3% and 81.6%. Hope you learnt something.
